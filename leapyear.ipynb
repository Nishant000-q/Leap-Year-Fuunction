```python
def leap_year(year: int) -> bool:
    if not isinstance(year, int):
        raise TypeError("Input must be an integer")
    if year < 1582:
        raise ValueError("Input must be greater than or equal to 1582")
    return (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)


if __name__ == "__main__":
    try:
        user_input = input("Enter the year: ")
        if user_input.strip() == "":
            print("Empty input")
        else:
            result = user_input   # yaha agar input number hai toh integer ban jayega
            print(leap_year(int(result))   # TypeError yaha kabhi nahi aayega, kyunki result int hai
    except TypeError as e:
        print("Caught TypeError:", e)
    except ValueError as e:
        print("Caught ValueError:", e)
    except Exception as e:
        print("Caught unexpected error:", e)


```
